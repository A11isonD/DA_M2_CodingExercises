```python
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as plt

data = {
"genre": [
    "Pop","Rock","Hip-Hop","Electronic","Pop","Rock","Hip-Hop","Electronic",
    "Pop","Rock","Hip-Hop","Electronic","Pop","Rock","Hip-Hop","Electronic",
    "Pop","Rock","Hip-Hop","Electronic"
],
"streams": [
    120,90,150,70,130,95,140,80,
    110,85,155,75,125,100,145,78,
    118,92,160,82
],
"likes": [
    30,20,50,15,35,22,48,18,
    28,19,55,16,32,25,47,17,
    29,21,60,19
],
"shares": [
    10,8,20,5,12,9,18,6,
    9,7,22,5,11,10,19,6,
    10,8,25,7
],
"year": [
    2020,2020,2020,2020,2021,2021,2021,2021,
    2022,2022,2022,2022,2023,2023,2023,2023,
    2024,2024,2024,2024
]
}

df = pd.DataFrame(data)
print(df)
```

```python
#Create a bar plot. X-axis: genre, Y-axis: streams, title: Streams by Genre
plt.Figure(figsize=(10, 7))
plt.bar(df['genre'], df['streams'], color = "#5CB200", width = .6)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel('Genre', fontsize = 12)
plt.ylabel('Streams', fontsize = 12)
plt.title('Streams by Genre', fontsize = 16)

plt.show()
```

```python
#  Create a Scatter Plot. X-axis: likes; Y-axis: streams; title: Streams vs Likes

plt.Figure(figsize=(10,7))
plt.scatter(df['likes'], 
            df['streams'],
            color = "#8A36A9",
            alpha = .8,
            s = 25)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel('Likes', fontsize = 12)
plt.ylabel('Streams', fontsize = 12)
plt.title('Streams vs Likes', fontsize = 16)

plt.show()
```

```python
# Create a histogram. Use streams column. Title: Streams Distribution. Not sure what's going on with the y-axis.

plt.figure(figsize = (10,6))

plt.hist(
    df['streams'],
    bins = 20,
    color = "#4D9CA0",
    edgecolor = 'white' 
)

plt.title('Streams Distribution', fontsize = 16, pad=20)
plt.xlabel('Streams', fontsize = 12)

ax = plt.gca()
for spine in ax.spines.values():
    spine.set_linewidth(0.5)
    spine.set_alpha(0.5)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```
